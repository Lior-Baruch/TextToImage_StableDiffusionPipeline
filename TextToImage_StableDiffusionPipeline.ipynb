```python
import torch
from diffusers import StableDiffusionPipeline

# set model id and device
model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"

# load model
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to(device)
```

```python
# create prompt, forward pass and plot result
import matplotlib.pyplot as plt
import numpy as np
prompt = "an image of an underwater world with colorful coral reefs and schools of fish."
image = pipe(prompt).images[0]

plt.figure(figsize=(10, 10))
plt.imshow(np.array(image))
plt.axis("off")
plt.show()
```

```python
# some examples
prompts = [
    "a picture of a cat",
    "a picture of a dog",
    "a picture of a person",
    "a picture of a car",
    "a picture of a house",
    "a picture of a tree",
    "a picture of a flower",
    "a picture of a bird",
    "a picture of a fish",
    "a picture of a mountain",
    "a picture of a lake",
    "a picture of a river",
    "a picture of a sunset",
    "a picture of a sunrise",
    "a picture of a rainbow",
    "a picture of a cloud",
    "a picture of a storm",
    "a picture of a lightning",
    "a picture of a snowflake",
    "a picture of a snowman",
    "animals in a zoo",
    "animals in a forest",
    "animals in a desert",
    "animals in a jungle"]

# compute images and plot, 4 images per row
plt.figure(figsize=(20, 20))
for i, prompt in enumerate(prompts):
    image = pipe(prompt).images[0]
    plt.subplot(6, 4, i + 1)
    plt.imshow(np.array(image))
    plt.axis("off")
    plt.title(prompt)
plt.show()
```
